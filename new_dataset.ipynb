````python
from datasets import load_dataset
import pandas as pd
import re

# 2. Load the dataset in "streaming" mode 
# Streaming means it won't download the full 347MB, it just streams the data you ask for
dataset = load_dataset("jtatman/python-code-dataset-500k", split="train", streaming=True)

# 3. Grab the first 5 records
first_few_records = list(dataset.take(5))

# 4. Convert to a Pandas DataFrame for a clean, tabular view in Jupyter
# Prevent pandas from truncating long text in the table cells
# pd.set_option("display.max_colwidth", None)
# df = pd.DataFrame(first_few_records)
# display(df)

# 5. Print the exact text of the very first record to see the structure
# for i in range(1):
#     print(f"\n--- RAW TEXT OF RECORD {i} ---")
#     print(first_few_records[i]['instruction']) 
#     print(f"\n--- CODE RESPONSE ---")
#     print(first_few_records[i]['output'])
````

````python
def extract_python_code(example):
    text = example['output']
    pattern = r'```(?:python)?\s*(.*?)\s*```'
    match = re.search(pattern, text, flags=re.DOTALL | re.IGNORECASE)
    
    if match:
        example['output'] = match.group(1).strip()
    else:
        example['output'] = ""
        
    return example

# 3. Apply the extraction to just this tiny test dataset
clean_test_dataset = list(map(extract_python_code, first_few_records))

# 4. Print the results to verify
for i, row in enumerate(clean_test_dataset):
    print(f"--- ROW {i} ---")
    print(row['output'])
    print("\n" + "="*40 + "\n")
````

````python
import dis

# 1. Define your Python source code as a string
source_code = """
def calculate_total(price, tax_rate):
    total = price + (price * tax_rate)
    return total

calculate_total(100, 0.05)
"""

# 2. Compile the source code into a bytecode object
# Arguments: source string, filename (can be '<string>'), and mode ('exec' for multiple lines)
compiled_bytecode = compile(source_code, '<string>', 'exec')

print("--- Compiled Bytecode Object ---")
print(compiled_bytecode)
print("\n--- Human-Readable Instructions (Disassembly) ---")

# 3. Disassemble the bytecode object into readable instructions
dis.dis(compiled_bytecode)
````
